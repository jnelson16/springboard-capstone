### Scatterplots and Regressions

#### Do Higher Team Salaries Translate to More Wins?

```python
team_salary = team.merge(salary)
```

```python
def wins_salary_graph(year):
    wins, salary_total = salary_year(year)
    team_salary_year = team_salary[team_salary.year == year]
    ws_win = team_salary_year[team_salary_year.ws_win == 'Y']
    plt.scatter(np.log(salary_total), wins)
    plt.scatter(np.log(ws_win.groupby('team_id')['salary'].sum()), ws_win.groupby('team_id')['w'].median())
    plt.ylabel("Team wins")
    plt.xlabel("Log team salary")
    plt.title(f"How does salary impact wins?\n{year} season")
```

```python
wins_salary_graph(2015)
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
def salary_year(year):
    team_salary_year = team_salary[team_salary.year == year]
    wins = team_salary_year.groupby('team_id')['w'].median()
    salary_total = team_salary_year.groupby('team_id')['salary'].sum()
    return wins, salary_total
```

```python
for year in range(1985, 2016):
    y, X = salary_year(year)
    results = sm.OLS(y, np.log(X)).fit()
    print(f'For {year}:\n{results.summary()}')
```

```python
wins_salary_graph(1985)
```

```python
wins_salary_graph(1987)
```

#### What about change in salary?

```python
def salary_change(year):
    team_salary_year = team_salary[team_salary.year == year]
    team_salary_minusone = team_salary[team_salary.year == (year - 1)]
    wins = team_salary_year.groupby('team_id')['w'].median()
    salary_total = team_salary_year.groupby('team_id')['salary'].sum() - team_salary_minusone.groupby('team_id')['salary'].sum()
    return wins.values.reshape(-1, 1), salary_total.values.reshape(-1, 1)
```

```python
for year in range(1986, 2016):
    try:
        X, y = salary_change(year)
        results = sm.OLS(y, np.log(X)).fit()
        print(f'For {year}: {results.summary()}')
    except ValueError:
        pass
```

```python
def wins_salary_change_graph(year):
    team_salary_year = team_salary[team_salary.year == year]
    team_salary_minusone = team_salary[team_salary.year == (year - 1)]
    wins = team_salary_year.groupby('team_id')['w'].median()
    salary_change = (
        (team_salary_year.groupby('team_id')['salary'].sum() - team_salary_minusone.groupby('team_id')['salary'].sum())
        / team_salary_minusone.groupby('team_id')['salary'].sum()
    )
    plt.scatter(salary_change*100, wins)
    plt.ylabel("Team wins")
    plt.xlabel("Team salary % change")
    plt.title(f"How does salary impact wins?\n{year} season")
```

```python
wins_salary_change_graph(2006)
```
